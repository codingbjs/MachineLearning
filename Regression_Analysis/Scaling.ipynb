```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns

DF = sns.load_dataset('mpg')
DF.info()
```

```python
DF.head()
```

```python
X = DF[['weight']]
y = DF['mpg']

X[:5]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (9,6))
sns.regplot(x=X,y=y)
plt.xlabel('weight_without_scaling')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(X)

X_Norm[:5]
```

```python
plt.figure(figsize = (9,6))
sns.regplot(x=X_Norm,y=y)
plt.xlabel('weight_without_normalization')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_Stan = scaler2.fit_transform(X)

X_Stan[:5]
```

```python
plt.figure(figsize = (9,6))
sns.regplot(x=X_Stan,y=y)
plt.xlabel('weight_without_StandardScaler')
plt.show()
```
