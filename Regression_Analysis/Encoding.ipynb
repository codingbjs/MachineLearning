```python
X = DF[['origin']]
X[111:115]
```

```python
from sklearn.preprocessing import LabelEncoder

encoder1 = LabelEncoder()
LE = encoder1.fit_transform(X)
LE[111:115]
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder2 = OneHotEncoder()
OHE = encoder2.fit_transform(X)
print(OHE[111:115])
```
